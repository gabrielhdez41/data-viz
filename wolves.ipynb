```python
import numpy as np
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt
import seaborn as sns
import squarify
```

```python
df_adv = pd.read_csv('~/Documentos/MEGA/wolves_adv.csv')
df_tot = pd.read_csv('~/Documentos/MEGA/wolves_tot.csv')
pd.set_option('display.max_columns', None)
```

```python
df_adv.drop(['Rk', 'Age', 'G','MP', 'Unnamed: 17', 'Unnamed: 22'], axis=1, inplace=True)
df_adv.rename({'Unnamed: 1': 'Name'}, axis=1, inplace=True) 
df_adv.columns = df_adv.columns.str.lower()
```

```python
df_tot.drop('Rk', axis=1, inplace=True)
df_tot.rename({'Unnamed: 1': 'Name'}, axis=1, inplace=True)
df_tot.columns = df_tot.columns.str.lower()
```

```python
df_adv.fillna(0, inplace=True)
df_tot.fillna(0, inplace=True)
```

```python
left = df_adv.set_index('name')
right = df_tot.set_index('name')

df_all = left.join(right, lsuffix='_adv', rsuffix='_tot')
```

```python
df_all
```

```python
x = df_all['pts']
y = df_all['ast']
n = df_all.index
plt.figure(figsize=(16,8))
sns.regplot(x=x,y=y)
plt.title('Puntos vs Asistencias (Wolves 20/21)', fontweight='bold')
for i, label in enumerate(n):
    plt.annotate(label, (x[i], y[i]))
```

```python
plt.figure(figsize=(16,8))
plt.pie(df_all['ast%'],labels = df_all.index)
plt.title("Wolves Assist")
plt.tight_layout()
```

```python
df_pts = df_all[df_all["pts"]>0]
#Utilise matplotlib to scale our goal numbers between the min and max, then assign this scale to our values.
norm = matplotlib.colors.Normalize(vmin=min(df_pts['pts']), vmax=max(df_pts['pts']))
colors = [matplotlib.cm.Blues(norm(value)) for value in df_pts['pts']]

#Create our plot and resize it.
fig = plt.gcf()
ax = fig.add_subplot()
fig.set_size_inches(16, 4.5)

#Use squarify to plot our data, label it and add colours. We add an alpha layer to ensure black labels show through
squarify.plot(label=df_pts.index,sizes=df_pts['pts'], color = colors, alpha=.6)
plt.title("Wolves Points",fontsize=23,fontweight="bold")

#Remove our axes and display the plot
plt.axis('off')
plt.show()
```
